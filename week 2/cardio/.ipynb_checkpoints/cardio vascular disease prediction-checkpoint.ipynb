## Weight is not an accurate measurement for cardiovascular disease so calculating bmi

BMI= weight(kg)/height(m)*2


```python
df["bmi"]=df["weight"]/((df["height"]/100)**2)
df.head()
```

## Using iqr method to remove outliers

```python
def remove_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

df = remove_outliers_iqr(df, 'age')
df = remove_outliers_iqr(df, 'ap_hi')
df = remove_outliers_iqr(df, 'ap_lo')
df = remove_outliers_iqr(df, 'bmi')
```

## Fitting classification models

```python
df.columns
```

### Logistic Regression

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
X = df[['age', 'gender', 'ap_hi', 'ap_lo', 'cholesterol', 'gluc', 'smoke','alco', 'active','bmi']]
y = df['cardio']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=40)
model=LogisticRegression(max_iter=100000)
model.fit(X_train,y_train)
y_pred = model.predict(X_test)
from sklearn.metrics import (accuracy_score,
                             f1_score,
                             confusion_matrix,
                             precision_score,
                             ConfusionMatrixDisplay,
                             recall_score,
                             classification_report)
y_pred = model.predict(X_test)

accuracy=accuracy_score(y_true=y_test, y_pred= y_pred)
precision=precision_score(y_true=y_test, y_pred= y_pred)
recall=recall_score(y_true=y_test, y_pred= y_pred)
f1=f1_score(y_true=y_test, y_pred= y_pred)
confusion_matrix=confusion_matrix(y_true=y_test, y_pred= y_pred)
report=classification_report(y_true=y_test, y_pred= y_pred)
print(f'accuracy={accuracy}')
print(f'precision={precision}')
print(f'recall={recall}')
print(f'f1 score={f1}')
print(report)
```

```python
ConfusionMatrixDisplay.from_estimator(model,X_test,y_test)
plt.show()
```

### Support Vector Classifier (SVC)

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=40)
from sklearn.svm import SVC
svc=SVC()
svc.fit(X_train,y_train)
y_pred = svc.predict(X_test)

accuracy=accuracy_score(y_true=y_test, y_pred= y_pred)
precision=precision_score(y_true=y_test, y_pred= y_pred)
recall=recall_score(y_true=y_test, y_pred= y_pred)
f1=f1_score(y_true=y_test, y_pred= y_pred)
report=classification_report(y_true=y_test, y_pred= y_pred)
print(f'accuracy={accuracy}')
print(f'precision={precision}')
print(f'recall={recall}')
print(f'f1 score={f1}')
print(report)
```

```python
ConfusionMatrixDisplay.from_estimator(svc,X_test,y_test)
plt.show()
```
